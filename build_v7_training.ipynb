```python
import pandas as pd
filtered_file_path = 'data/v7/filter_v7/filtered_split_pair_all_siamese_bert_base_cased_epoch2.csv'
filtered_pairs = pd.read_csv(filtered_file_path, sep='\t')
print(len(filtered_pairs))
filtered_pairs.head()
```

```python
original_pair_path = 'data/pair_v7/paired_sentences_train.csv'
original_pairs = pd.read_csv(original_pair_path, header=None, sep='\t')
print(len(original_pairs))
original_pairs.head()
```

```python
import numpy as np

rs = np.concatenate([original_pairs.values, filtered_pairs.values])
updated_original = pd.DataFrame(rs, columns=original_pairs.columns)
print(len(updated_original))
updated_original.tail(10)
```

```python
new_dataset_file_path = 'data/pair_v7/paired_sentences_train_v7.2.csv'
updated_original.to_csv(new_dataset_file_path, index=False, sep='\t', header=None)
```
