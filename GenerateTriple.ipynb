```python
import pandas as pd 
import numpy as np 

SAMPLE_SIZE = 1200000   # 100w positive, 100w negative

train = pd.read_csv("./data/train_set_v1.txt", sep='\t', header=None)
```

```python
def generateSentencePair(sentence_list):
    sentence_pair_list = []
    if len(sentence_list) <= 1:
        return sentence_pair_list
    for i in range(len(sentence_list)):
        for j in range(i+1, len(sentence_list)):
#             a = np.random.randint(low=0, high=len(train), size=2)
            
            sentence_pair_list.append([sentence_list[i], sentence_list[j]])
    return sentence_pair_list
```

```python
id_list = np.array(list(train[0]))
sentence_list = np.array(list(train[1]))
id_unique_list = np.unique(id_list)

positive_pairs = []

for i in id_unique_list:
    index_list = (id_list == i)
    id_list[index_list]
    sentence_pairs = generateSentencePair(sentence_list[index_list])
    positive_pairs.extend(sentence_pairs)
len(positive_pairs)
```

```python
choices = np.random.choice(len(positive_pairs), SAMPLE_SIZE, replace=False)
positive_pairs_np = np.array(positive_pairs)
filtered_positive_pairs = positive_pairs_np[choices]
```

```python
filtered_positive_pairs.shape
```

```python
choices = np.random.choice(len(train), SAMPLE_SIZE, replace=True)
negative_examples = np.array(train[1])[choices]
```

```python
all_triple = np.concatenate((filtered_positive_pairs, negative_examples.reshape(len(negative_examples), 1)), axis=1)
```

```python
all_triple.shape
```

```python
triple_df = pd.DataFrame(all_triple)
triple_df
```

```python
triple_df.to_csv('triple_sentences.csv', sep='\t', header=None, index=False)
```
